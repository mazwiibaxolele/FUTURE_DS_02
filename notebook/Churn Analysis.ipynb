```python
import pandas as pd
```

```python
df = pd.read_csv('exports/telco_churn_cleaned.csv')
```

```python
total_customers = df.shape[0]
print(f"Total Customers: {total_customers}")
```

```python
churned_customers = df[df['Churn']=='Yes'].shape[0]
print(f"Churned Customers: {churned_customers}")
```

```python
churn_rate = (churned_customers/total_customers)*100
print(f"Churn Rate: {round(churn_rate, 2)}%")
```

```python
retention_rate = 100-churn_rate
print(f"Retention Rate: {round(retention_rate, 2)}%")
```

```python
contract_churn = pd.crosstab(df['Contract'], df['Churn'])
print(contract_churn)
```

```python
contract_churn['Total'] = contract_churn['No'] + contract_churn['Yes']
contract_churn['Churn Rate (%)'] = round(
    contract_churn['Yes'] / contract_churn['Total'] * 100, 2
)
print(contract_churn)
```

```python
df['TenureGroup'] = pd.cut(
    df['tenure'],
    bins=[0, 12, 24, 48, 72],
    labels=["0-1 Year", "1-2 Years", "2-4 Years", "4-6 Years"] 
)
```

```python
tenure_churn = pd.crosstab(df['TenureGroup'], df['Churn'])
tenure_churn['Total'] = tenure_churn['No'] + tenure_churn['Yes']
tenure_churn['Churn Rate (%)'] = round(
    tenure_churn['Yes'] / tenure_churn['Total'] * 100, 2
)
print(tenure_churn)
```

```python
monthly_charges = df.groupby('Churn')['MonthlyCharges'].mean()
print(f"Average Monthly Charges: {monthly_charges}")
```

```python
total_charges = df.groupby('Churn')['TotalCharges'].mean()
print(f"Average Total Charges: {total_charges}")
```
